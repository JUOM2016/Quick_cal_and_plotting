```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
```

```python
dir = 'C:\\Users\\wj2002\\Dropbox (Heriot-Watt University Team)\\RES_EPS_Quantum_Photonics_Lab\\Experiments\\Current Experiments\\Broadband telecom quantum memories\\1550_cavity_locking'
folder = '20221129'
```

```python
plot_data=['1550C-Error_T.CSV']
path_data= dir + '\\' + folder + '\\' + plot_data[0]
data=pd.read_csv(path_data)
data=data.to_numpy() # Read the raw data when the laser is focused into the sample

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(data[:,0], data[:,2], 'r-', label='Error signal')
ax1.plot(data[:,0], data[:,3], 'b-', label='Transmitted signal')
ax2.plot(data[:,0], data[:,1], 'k-', label='Piezo signal')
fig.legend()
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude (V)')
ax2.set_ylabel('Piezo signal')
ax1.set_ylim([-0.3,0.46])
ax2.set_ylim([0.3,0.53])
fig.patch.set_facecolor('white')

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(data[:,0], data[:,2], 'r-', label='Error signal')
ax2.plot(data[:,0], data[:,3], 'b-', label='Transmitted signal')
fig.legend()
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Error signal amplitude (V)', color='r')
ax2.set_ylabel('Transmitted signal amplitude (V)', color='b')
ax1.set_xlim([-0.3,0])
ax1.set_ylim([-0.3,0.46])
ax2.set_ylim([0,0.3])
fig.patch.set_facecolor('white')
```
